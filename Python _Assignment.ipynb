6. Write a function to demonstrate the behavior of mutable and immutable arguments.

```python
def modify_values(num, num_list):
    num += 10  # Immutable argument
    num_list.append(100)  # Mutable argument

# Original values
a = 5
b = [1, 2, 3]

# Before function call
print("Before:", a, b)

# Call function
modify_values(a, b)

# After function call
print("After:", a, b)  # 'a' doesn't change, but 'b' does
```

9. Write a Python program to convert user input from string to integer, float, and boolean types.

```python
# Get user input as a string
user_input = input("Enter a value: ")

# Convert to different types
integer_value = int(user_input) if user_input.isdigit() else "Invalid integer"
float_value = float(user_input) if user_input.replace('.', '', 1).isdigit() else "Invalid float"
boolean_value = user_input.lower() in ['true', '1', 'yes']

# Print the converted values
print("Integer:", integer_value)
print("Float:", float_value)
print("Boolean:", boolean_value)
```

10. Write code to demonstrate type casting with list elements.

```python
# Original list with different data types (string, int, float)
mixed_list = ["123", "45.67", 89, 12.34]

# Convert each element to an integer, float, or string
int_list = [int(float(x)) if isinstance(x, (str, float)) and str(x).replace('.', '', 1).isdigit() else x for x in mixed_list]
float_list = [float(x) if isinstance(x, (str, int)) else x for x in mixed_list]
str_list = [str(x) for x in mixed_list]

# Print the results
print("Original List:", mixed_list)
print("List after converting to integers:", int_list)
print("List after converting to floats:", float_list)
print("List after converting to strings:", str_list)
```

13. Write a Python program to find the sum of all even numbers between 1 and 50.

```python
# Initialize the sum
sum_of_evens = 0

# Loop through numbers from 1 to 50
for num in range(1, 51):
    if num % 2 == 0:
        sum_of_evens += num

# Print the result
print("Sum of all even numbers between 1 and 50:", sum_of_evens)
```

14. Write a program to reverse a string using a while loop.

```python
# Input string from user
string = input("Enter a string: ")

# Initialize an empty string to store the reversed string
reversed_string = ""

# Initialize an index to start from the last character of the string
index = len(string) - 1

# Use a while loop to reverse the string
while index >= 0:
    reversed_string += string[index]  # Add the character to the reversed string
    index -= 1  # Move to the previous character

# Print the reversed string
print("Reversed string:", reversed_string)
```

15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.

```python
# Input number from user
num = int(input("Enter a number: "))

# Initialize factorial variable
factorial = 1

# Calculate factorial using a while loop
while num > 1:
    factorial *= num
    num -= 1

# Print the result
print("The factorial is:", factorial)
```
